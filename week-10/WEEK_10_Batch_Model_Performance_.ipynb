```python
import matplotlib.pyplot as plt
import numpy as np

# Define result_scores as a dictionary of lists for each step
result_scores = {
    'Dummy': [0.50, 0.52, 0.49],
    'KNN-1': [0.75, 0.76, 0.74],
    'KNN-2': [0.70, 0.72, 0.68],
    'KNN-3-1': [0.78, 0.79, 0.77],
    'KNN-3-2': [0.85, 0.88, 0.84]
}

# Data from result_scores
steps = ['Dummy', 'KNN-1', 'KNN-2', 'KNN-3-1', 'KNN-3-2']
metrics = ['Accuracy', 'Recall', 'Precision']

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot each metric
x = np.arange(len(steps))
width = 0.25

for i, metric in enumerate(metrics):
    # Replace 'data' with 'result_scores' to access the data
    values = [result_scores[step.split()[0]][i] for step in steps]
    ax.bar(x + i*width, values, width, label=metric)

# Add labels, title and legend
ax.set_xlabel('Model Steps')
ax.set_ylabel('Score')
ax.set_title('Model Performance Across Batch Steps')
ax.set_xticks(x + width)
ax.set_xticklabels(steps)
ax.legend()

# Add value labels on top of bars
for i, step in enumerate(steps):
    for j, metric in enumerate(metrics):
        # Replace 'data' with 'result_scores' to access the data
        value = result_scores[step.split()[0]][j]
        ax.text(i + j*width, value + 0.01, f'{value:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

INTERPRETATION: The bar chart titled **Model Performance Across Batch Steps** illustrates how different model configurations perform across various stages—namely Dummy, KNN-1, KNN-2, KNN-3-1, and KNN-3-2—based on three key evaluation metrics: Accuracy, Recall, and Precision. The Dummy model, likely a baseline classifier, performs close to chance level with an accuracy of 0.50, recall of 0.52, and precision of 0.49. A significant improvement is seen in the KNN-1 model, which achieves balanced performance across all metrics, indicating a well-calibrated configuration. However, KNN-2 experiences a slight performance dip across the board, suggesting that this iteration may involve suboptimal parameter tuning or experimental variation. The trend picks up again with KNN-3-1, which slightly outperforms both KNN-1 and KNN-2. The best performance is observed with KNN-3-2, boasting an accuracy of 0.85, recall of 0.88, and precision of 0.84, highlighting its superior reliability and robustness. Overall, the chart reflects a positive progression in model effectiveness, with KNN-3-2 emerging as the optimal configuration due to its consistently high scores across all metrics.
